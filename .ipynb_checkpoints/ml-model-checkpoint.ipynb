```python
import pandas as pd

def getDataFromSheet(file_path):
    try:
        # Read the Excel file into a pandas DataFrame
        data = pd.read_excel(file_path, usecols='A')  # Use only column A

        # Extract the values from the DataFrame into a list
        ans = data['A'].tolist()
        print(ans)
        return ans
    except Exception as e:
        print(f"An error occurred: {str(e)}")
        return None
```
